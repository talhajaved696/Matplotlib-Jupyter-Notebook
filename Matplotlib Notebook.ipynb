## Matlpotlib Basics 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


#Line1
plt.plot(x,y,label='2x',color='red',linewidth='2',marker='o',markersize=10,markerfacecolor='blue',markeredgecolor='red')
#short code [color][marker][linestyle]
#plt.plot(x,y,'r*--',label='2x')
#Line 2
x2 = np.arange(0,4,0.5)
#plt.plot(x2,x2**2,'ro--',label='X^2',markersize=5,markeredgecolor='b')
plt.plot(x2[0:6],x2[:6]**2,'r*--',label='X^2',markersize=5,markeredgecolor='b')
plt.plot(x2[5:],x2[5:]**2,'ro-',label='X^2',markersize=5,markeredgecolor='b')
plt.title('Line Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.grid()
plt.savefig('linegraph.png',dpi=300)
#plt.legend(loc='upper center', bbox_to_anchor=(1.45, 0.8), shadow=True, ncol=1) legend outside
plt.legend(loc='upper left')
plt.show()
```

### Bar Chart

```python
plt.style.use('default')
labels = ('A', 'B', 'C')
y = np.arange(len(labels))
print(y,labels)
values = [1,4,2]


bars = plt.bar(y, values,align='center',alpha=1)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.xticks(y,labels)
plt.ylabel('Values')
plt.title('Bar Chart',fontdict={'fontname':'Comic Sans MS','fontsize':20})
#plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
fifa = pd.read_csv('D:\\matplotlib_data\\fifa_data.csv')
fifa
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018',fontdict={'fontweight':'bold','fontsize':15})
plt.xticks(bins)
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left','Preferred Foot'].count()
right = fifa.loc[fifa['Preferred Foot'] == 'Right','Preferred Foot'].count()
plt.pie([left,right],labels=['Left Foot','Right Foot'],colors=['#abcdef','blue'],autopct='%.2f %%',shadow=True)

plt.title('Foot Preference of FIFA Players')
#plt.axis('equal')
plt.show()

```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125,'Weight'].count()
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150),'Weight'].count()
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175),'Weight'].count()
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200),'Weight'].count()
heavy = fifa.loc[fifa.Weight >= 200,'Weight'].count()

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')
#plt.axis('equal')
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%',startangle=180)
plt.show()
```

```python
plt.figure(figsize=(5,7), dpi=100)
plt.style.use('default')

barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
realmadrid = fifa['Overall'][fifa['Club'] == 'Real Madrid']
revs = fifa.loc[fifa.Club == "New England Revolution",'Overall']

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
labels = ['FC Barcelona', 'Real Madrid','England Revo']
bp = plt.boxplot([barcelona, realmadrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})
for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')
plt.show()
```
